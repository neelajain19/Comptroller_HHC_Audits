```python
## import libraries

import regex as re
import pandas as pd
import glob
from mangoCR import pdf2image_ocr_text
from rapidfuzz import fuzz
from functools import lru_cache
import spacy
```

```python
## Comptroller Report PDFs downloaded as part of Assignment 1
## saved PDFs in folder (Comptroller_Reports) next to this nwhat is globbing pdfotebook
## glob PDFs together
target_pdfs = glob.glob("Comptroller_Reports/*pdf")
target_pdfs
```

```python
## Create a list of search terms for your pdfs

keywords = ["Coney Island Hospital",
            "Brooklyn Health South",
            "improper billing practices",
            "Medicaid reimbursement",
            "Managed Care Organization",
            "CSC billing policies",
            "Fee for Service",
            "(FFS)", 
            "Anthony Rajkumar", 
            "William Brown",
            "Svetlana Lipyanskaya",]
```

```python
## Read through text of all 20 PDF files and store results in a datadframe
df_list = []
for i, target_pdf in enumerate(target_pdfs, start = 0):
    print(f"Processing {i} of {len(target_pdfs)}... ")
    text = pdf2image_ocr_text(target_pdf)  ## converts to text
    print(f"Looking for context words... ")
    df_list.append(keyword_context_finder(
    content=text,
    terms=keywords,
    file_name= target_pdf,
    words_around=15)) ## indicates the number of words to display for context
```

```python
## confirm df list contains all PDFs
## how long is df_list?

len(df_list)
```

```python
## concat together for ease
df = pd.concat(df_list, ignore_index= True)
```

```python
## preview df
df.sample(20)
```

```python
## save df to csv to look at results better
df.to_csv('Comptroller_Report_KW_ID.csv', index=False) 
```
